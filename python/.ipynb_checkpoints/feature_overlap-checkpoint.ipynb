```python
# The goal of this project is to create a bedtools-like program which identifies the intersection of 
# ChIP-seq peaks with annotated enhancers regions.

from pathlib import Path
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib_venn import venn3
%matplotlib inline
```

```python
#Load three datasets. EP300 and CTCF files contain ChIP-seq peaks for the transcription factors EP300 and CTCF.
#The K562 file contains a set of annotated enhancers from the K562 cell line. 
EP300_PATH = Path("../src/EP300_ChIP-seq_peaks.bed")
CTCF_PATH = Path("../src/CTCF_ChIP-seq_peaks.bed")
K562_PATH = Path("../src/K562-Enhancers.bed")

def load_df(path):
    return pd.read_csv(path, sep='\t', header=None).sort_values(by=[0, 1])

EP300_df = load_df(EP300_PATH)
CTCF_df = load_df(CTCF_PATH)
K562_df = load_df(K562_PATH)
```

```python
#Takes two lists (containing both start and stop coordinates) and iterates through identifying overlapping regions
def find_intersection(dataset1, dataset2):
    intersection = []
    for range1 in dataset1:
        for range2 in dataset2:
            start = max(range1[0], range2[0])
            end = min(range1[1], range2[1])
            if start <= end:
                intersection.append((start, end))
    return intersection


CHROM = ['chr1', 'chr2', 'chr3', 'chr4', 'chr5', 'chr6', 'chr7',
       'chr8', 'chr9', 'chr10', 'chr11', 'chr12', 'chr13', 'chr14',
       'chr15', 'chr16', 'chr17', 'chr18', 'chr19', 'chr20', 'chr21', 'chr22',
       'chrX']
#Takes two (optionally three) dataframes and iterates through each chromosome and inputs it into the find_intersection function
#to identify overlapping regions.
def intersect_by_chrom(dataset1, dataset2, dataset3=None):
    chrom_intersection = []
    for group in CHROM:
        filtered_df1 = dataset1[dataset1[0] == group].iloc[:, [1, 2]]
        filtered_df2 = dataset2[dataset2[0] == group].iloc[:, [1, 2]]
        
        overlap_list = find_intersection(filtered_df1.values.tolist(),
                                          filtered_df2.values.tolist())

        if dataset3 is not None:
            filtered_df3 = dataset3[dataset3[0] == group].iloc[:, [1, 2]]
            overlap_list = find_intersection(overlap_list,
                                             filtered_df3.values.tolist())

        chrom_intersection.append(overlap_list)

    return chrom_intersection
```

```python
#Printed output of the sum of intersections between ranges.
print('K562: '+str(len(K562_df)))
print('EP300: '+str(len(EP300_df)))
print('CTCF: '+str(len(CTCF_df)))
output = intersect_by_chrom(K562_df,EP300_df)
K562_EP300_intersect = sum(len(sublist) for sublist in output)
print('K562_EP300: '+str(K562_EP300_intersect))
output2 = intersect_by_chrom(K562_df,CTCF_df)
K562_CTCF_intersect = sum(len(sublist) for sublist in output2)
print('K562_CTCF: '+str(K562_CTCF_intersect))
output3 = intersect_by_chrom(EP300_df,CTCF_df)
EP300_CTCF_intersect = sum(len(sublist) for sublist in output3)
print('EP300_CTCF: '+str(EP300_CTCF_intersect))
output4 = intersect_by_chrom(EP300_df,CTCF_df,K562_df)
triple_intersect = sum(len(sublist) for sublist in output4)
print('triple intersect: '+str(triple_intersect))
```

```python
#Visualization of intersections
v=venn3(subsets = (len(K562_df),
                   len(EP300_df),
                   K562_EP300_intersect,
                   len(CTCF_df),
                   K562_CTCF_intersect,
                   EP300_CTCF_intersect,
                   triple_intersect), set_labels = ('K562_Enhancers', 
                                                    'EP300 ChIP-seq peaks', 
                                                    'CTCF ChIP-seq peaks'))
plt.show()
```
